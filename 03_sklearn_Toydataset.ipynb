```python
# 모델 클래스 선택 → 학습기

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# 정답률, 에러률 임포트
from sklearn.metrics import accuracy_score, mean_squared_error
# 
from sklearn.model_selection import train_test_split
```

## 붓꽃 데이터 셋

```python
# 1) 데이터 생성

iris = load_iris()
```

```python
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris.sample(2)
```

```python
df_iris['label'] = iris.target
df_iris.sample(2)
```

```python
# 라벨 확인 

df_iris.label.values
```

```python
df_iris.info()
```

```python
# 2) 학습데이터, 데스트데이터 분리

X = df_iris.drop(columns=['label']).copy()
X
```

```python
y = df_iris['label']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, random_state=1)
```

```python
X_train.shape , X_test.shape 
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
# 3) 모델 생성

model_knn = KNeighborsClassifier()
model_dt = DecisionTreeClassifier()
model_svc = SVC()
```

```python
model_knn.fit( X_train, y_train )   # 학습 시킴
model_dt.fit( X_train, y_train )
model_svc.fit( X_train, y_train )
```

```python
model_knn.score(X_test, y_test)
```

```python
model_dt.score(X_test, y_test)
```

```python
print(accuracy_score(y_test, model_knn.predict(X_test)))
print(accuracy_score(y_test, model_dt.predict(X_test)))
print(accuracy_score(y_test, model_svc.predict(X_test)))
```
